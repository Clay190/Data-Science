# Ballistics and the Range Equation

*This is an investigation of the so-called range equation:* $d=\frac {v^2} g\sin 2\theta$



```python
from math import radians, sin, pi
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
def rangeformula(v, theta):
    impact = (v**2)/(9.81)*sin(2*radians(theta))
    print("Your impact distance is", impact)
    return impact
```

*Please input rangeformula(v, theta)

**Inputs**
* v - velocity in m/s
* theta - launch angle with respect to horizon in degrees (will be converted to radians by program)

```python
angles = linspace(0, 90, num = 200)
rangefromdegrees = [rangeformula(100, radians(angle)) for angle in angles]
print(rangefromdegrees)
```

```python
rangefromdegrees200 = [rangeformula(200, radians(angle)) for angle in angles]
rangefromdegrees00 = [rangeformula(300, radians(angle)) for angle in angles]
```

```python
plt.plot(angles,rangefromdegrees)
plt.plot(angles,rangefromdegrees200)
plt.plot(angles,rangefromdegrees300)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show()
```
