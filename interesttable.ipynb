```python
from numpy import linspace
import matplotlib.pyplot as plt
import pandas as pd
```

```python
years = [x for x in linspace(0,30,31)]
value_float = [(x, 1000*(1.035**x)) for x in linspace(0,30,31)]
value_string = [(x, round(1000*(1.035**x))) for x in linspace(0,30,31)]
```

```python
print(value_string)
```

```python
df = pd.DataFrame()
df['years'] = years
df['value_float'] = value_float
df['value_string'] = str(value_string[0])
```

```python
df
```

```python
plt.plot(value_float)
plt.title('Value in Relation to Years')
plt.xlabel('Years')
plt.ylabel('Value $')
plt.show()
```
