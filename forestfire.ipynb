**Forest Fire Data Set Analysis**

This is a look at the relationship between temperature and the DMC on dates that forest fires occur. DMC is an acronym for The Duff Moisture Code. This is a numeric rating, produced by the Canadian Wildlife website, of the average moisture content of loosely compacted organic layers of moderate depth soil. Through this project, I wanted to see if there was a correlation between the moisture content of soil.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'

df = pd.read_excel("https://docs.google.com/spreadsheets/d/e/2PACX-1vSKNNYvfLyQQtaxtnAxvAymU1YBxWCsQJRIDFyFIJz_BGOfkM7JFLf6p1243y7NXa_lIwdr8Ia5KROZ/pub?output=xlsx")

plt.scatter(df.temp, df.DMC)
plt.title('Relationship Between Temperature and DMC')
plt.xlabel('Temperature (Degrees Celsius)')
plt.ylabel('DMC')
plt.show()
```

This graph demonstrates a slightly positive correlation between the temperatuer and moisture in soil. This is an expected result, however my result still intrigues me because by plotting it, we can see how drastic an increase in 5 degrees decreases the amount of moisture in the soil.
